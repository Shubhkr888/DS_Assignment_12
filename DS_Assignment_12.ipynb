```python
#Question1
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def deleteMiddleNode(head):
    if not head or not head.next:
        return None

    slow = head
    fast = head
    prev = None

    while fast and fast.next:
        fast = fast.next.next
        prev = slow
        slow = slow.next

    prev.next = slow.next

    return head


# Test the function
def printLinkedList(head):
    curr = head
    while curr:
        print(curr.val, end=" ")
        curr = curr.next
    print()


# Create the linked list: 1->2->3->4->5
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print("Original linked list:")
printLinkedList(head)

head = deleteMiddleNode(head)

print("Modified linked list:")
printLinkedList(head)

```

```python
#Question2
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def hasLoop(head):
    if not head or not head.next:
        return False

    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            return True

    return False


# Test the function
def createLinkedList(arr, x):
    head = ListNode(arr[0])
    curr = head
    loopNode = None

    for i in range(1, len(arr)):
        newNode = ListNode(arr[i])
        curr.next = newNode
        curr = newNode

        if i == x:
            loopNode = curr

    if loopNode:
        curr.next = loopNode

    return head


# Example test case
arr = [1, 3, 4]
x = 2
head = createLinkedList(arr, x)

print("Does the linked list have a loop?")
print(hasLoop(head))
```

```python
#Question3
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def findNthFromEnd(head, N):
    if not head:
        return None

    fast = head
    slow = head

    # Move the fast pointer N nodes ahead
    for _ in range(N):
        if not fast:
            return None
        fast = fast.next

    # Move both the fast and slow pointers simultaneously
    while fast:
        fast = fast.next
        slow = slow.next

    return slow.val


# Test the function
def createLinkedList(arr):
    head = ListNode(arr[0])
    curr = head

    for i in range(1, len(arr)):
        newNode = ListNode(arr[i])
        curr.next = newNode
        curr = newNode

    return head


# Example test case
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
N = 2
head = createLinkedList(arr)

print("Value of the Nth node from the end:")
print(findNthFromEnd(head, N))
```

```python
#Question4
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def isPalindrome(head):
    if not head or not head.next:
        return True

    slow = head
    fast = head

    # Move fast pointer to the end and slow pointer to the middle
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    # Reverse the second half of the linked list
    prev = None
    curr = slow
    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node

    # Move fast pointer back to the head
    fast = head

    # Compare values of nodes in first and second half
    while prev:
        if fast.val != prev.val:
            return False
        fast = fast.next
        prev = prev.next

    return True


# Test the function
def createLinkedList(arr):
    head = ListNode(arr[0])
    curr = head

    for i in range(1, len(arr)):
        newNode = ListNode(arr[i])
        curr.next = newNode
        curr = newNode

    return head


# Example test cases
arr1 = ['R', 'A', 'D', 'A', 'R']
head1 = createLinkedList(arr1)
print("Is the linked list a palindrome?")
print(isPalindrome(head1))

arr2 = ['C', 'O', 'D', 'E']
head2 = createLinkedList(arr2)
print("Is the linked list a palindrome?")
print(isPalindrome(head2))
```

```python
#Question5
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def detectAndRemoveLoop(head):
    if not head:
        return None

    slow = head
    fast = head

    # Find the meeting point of slow and fast pointers
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            break

    # No loop found, return the original linked list
    if fast is None or fast.next is None:
        return head

    # Move slow pointer back to the head
    slow = head

    # Move both pointers one step at a time until they meet again
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next

    # Remove the loop by setting next pointer of the node before meeting point to None
    fast.next = None

    return head


# Test the function
def createLinkedList(arr, x):
    head = ListNode(arr[0])
    curr = head
    loopNode = None

    for i in range(1, len(arr)):
        newNode = ListNode(arr[i])
        curr.next = newNode
        curr = newNode

        if i == x:
            loopNode = curr

    if loopNode:
        curr.next = loopNode

    return head


# Example test case
arr = [1, 3, 4]
x = 2
head = createLinkedList(arr, x)

print("Original linked list:")
curr = head
while curr:
    print(curr.val, end=" -> ")
    curr = curr.next
print("None")

head = detectAndRemoveLoop(head)

print("Linked list after removing the loop:")
curr = head
while curr:
    print(curr.val, end=" -> ")
    curr = curr.next
print("None")
```
